```python
import pandas as pd
import numpy as np
import glob
from ydata_profiling import ProfileReport
```

```python

# Obtener todos los CSV de staging
archivos_csv = sorted(glob.glob('../data/staging/*.csv'))

print(f"Archivos encontrados: {len(archivos_csv)}")
for archivo in archivos_csv:
    print(f"  - {archivo}")
```

```python
# Cargar todos los CSV en una lista de dataframes
dfs = []

for archivo in archivos_csv:
    df = pd.read_csv(archivo)
    dfs.append(df)
    print(f"{archivo.split('/')[-1]}: {len(df)} filas")

print(f"\nTotal de dataframes: {len(dfs)}")
```

```python
# Concatenar verticalmente
df_completo = pd.concat(dfs, axis=0, ignore_index=True)

print(f"Dataframe final: {df_completo.shape}")
print(f"Columnas: {df_completo.columns.tolist()}")
```
